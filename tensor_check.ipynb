```python
import copy
import numpy as np
import mlx
import mlx.nn as mx
import mlx.core as mx_core
import torch
import torch.nn as nn
import torch.nn.functional as F
import safetensors


# 테스트 모델 Gemma-2b-it 업로드
path = "./model/"
model_name = "model-{}-of-{}.safetensors"
model1 = safetensors.safe_open(path+model_name.format("00001", "00002"), framework="pt")
model2 = safetensors.safe_open(path+model_name.format("00002", "00002"), framework="pt")


# model1, model2의 tensor들을 담음
safe_tensors = {}
for key in model1.keys() + model2.keys():
    if key in model1.keys():
        safe_tensors[key] = model1.get_tensor(key).type(torch.float32)
    elif key in model2.keys():
        safe_tensors[key] = model2.get_tensor(key).type(torch.float32)

# torch.weight 업로드
weight = torch.load("model/gemma-1.1-2b-it.ckpt", mmap=True, weights_only=True)["model_state_dict"]
new_tensors = {}
for key in weight.keys():
    if "complex" in str(weight[key].dtype):
        continue
    new_tensors[key] = weight[key].type(torch.float32)
```
